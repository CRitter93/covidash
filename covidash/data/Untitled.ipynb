```python
import pandas as pd
import os.path

DATA_DIR = '../../data'


raw_data_df = pd.read_csv(os.path.join(DATA_DIR, 'raw', 'rki_covid_19.csv'))
landkreis_df = pd.read_csv(os.path.join(DATA_DIR, 'raw', 'rki_corona_landkreise.csv'))
covid_19_df = raw_data_df.merge(landkreis_df[['county', 'OBJECTID']], left_on='Landkreis', right_on='county')
covid_19_df
```

```python
covid_19_df.Meldedatum = covid_19_df.Meldedatum.astype('datetime64').dt.strftime('%Y-%m-%d')
```

```python
covid_19_df.columns
```

```python
covid_19_df.groupby(['OBJECTID', 'Meldedatum']).agg({'AnzahlFall': 'sum', 'AnzahlTodesfall': 'sum'}).reset_index().sort_values(['OBJECTID', 'Meldedatum'])
```
