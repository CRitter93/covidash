```python
covid_19 = pd.read_csv('../../data/processed/rki_covid_19.csv')
covid_19
```

```python
idx = pd.date_range(covid_19.date.min(), covid_19.date.max())
idx
```

```python
covid_19 = covid_19.set_index('date')
covid_19.index = pd.DatetimeIndex(covid_19.index)
```

```python
covid_19 = covid_19.groupby(['OBJECTID', 'age', 'gender', 'GEN']).apply(lambda x: x[['cases', 'deaths']].reindex(idx).fillna(0)).reset_index().rename(columns={'level_4': 'date'})
```

```python
covid_idx = covid_19.set_index('date')
```

```python
%time covid_idx[covid_idx.index == '2020-02-20']
```

```python
pd.options.display.max_rows = 800
dimensions = ['OBJECTID', 'age', 'gender', 'GEN']
covid_19[['total_cases', 'total_deaths']] = covid_19.groupby(dimensions)['cases', 'deaths'].cumsum()
```

```python
covid_19[['test_c', 'test_d']] = covid_19.groupby(dimensions)[['cases', 'deaths']].rolling(4, min_periods=1).mean().reset_index()[['cases', 'deaths']]
```

```python
covid_19[['test_c_', 'test_d_']] = covid_19.groupby(dimensions)[['test_c', 'test_d']].apply(lambda x: x[['test_c', 'test_d']] - x[['test_c', 'test_d']].shift(1)).reset_index()[['test_c', 'test_d']]
covid_19[covid_19.GEN == 'Hamburg']
```

```python
covid_19.groupby(dimensions)[['test_c', 'test_d']].apply(lambda x: x[['test_c']] / x[['test_c']].shift(1)).reset_index()
```
